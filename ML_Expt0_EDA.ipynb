```python
df = df.drop(columns=['Championship Years'])
df.info()
```

Statistical Measures

Data Visualization

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Years_Active'], bins=20)
plt.title('Distribution of Years Active for F1 Drivers')
plt.xlabel('Years Active')
plt.ylabel('Number of Drivers')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
championship_counts = df['Championships'].value_counts().sort_index()
sns.barplot(x=championship_counts.index, y=championship_counts.values)
plt.title('Number of Drivers by Championships Won')
plt.xlabel('Number of Championships')
plt.ylabel('Number of Drivers')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
top_nationalities = df['Nationality'].value_counts().nlargest(10).index
sns.countplot(y='Nationality', data=df[df['Nationality'].isin(top_nationalities)], order=top_nationalities)
plt.title('Top 10 Nationalities of F1 Drivers')
plt.xlabel('Number of Drivers')
plt.ylabel('Nationality')
plt.show()
```

```python
df_race_starters = pd.DataFrame(df.groupby('Driver')['Race_Starts'].sum().sort_values(ascending=False)).reset_index()

plt.figure(figsize = (12,6))
sns.barplot(x = 'Driver', y = 'Race_Starts',data = df_race_starters.head(10))
plt.title('Top 10 Drivers by Race Starts')
plt.xlabel('Drivers')
plt.ylabel('Number of Race Starts')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Race_Starts', y='Race_Wins', data=df)
plt.title('Race Starts vs Race Wins for F1 Drivers')
plt.xlabel('Race Starts')
plt.ylabel('Race Wins')
plt.show()
```
