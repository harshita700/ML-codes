```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
```

####Dataset-1:


```python
data = {'X': [2, 3, 4, 5, 6, 7, 8, 9, 10],
        'Y': [1, 3, 6, 9, 11, 13, 15, 17, 20]}
df = pd.DataFrame(data)

X = df['X']
Y = df['Y']
n = len(df)
```

```python
x_mean = np.mean(X)
y_mean = np.mean(Y)

numerator_beta1 = np.sum((X - x_mean) * (Y - y_mean))
denominator_beta1 = np.sum((X - x_mean)**2)
beta1 = numerator_beta1 / denominator_beta1

beta0 = y_mean - beta1 * x_mean

print(f"Beta_0 (Intercept): {beta0}")
print(f"Beta_1 (Slope): {beta1}")
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['X'], df['Y'])
plt.title('Scatter Plot of X vs. Y')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
y_predicted = beta0 + beta1 * X

MSE = np.sum((Y - y_predicted)**2) / n

RMSE = np.sqrt(MSE)

print(f"Mean Squared Error (MSE): {MSE}")
print(f"Root Mean Squared Error (RMSE): {RMSE}")
```

```python
x = df_salary['YearsExperience'].values
y = df_salary['Salary'].values
```

```python
df_salary.describe()
```

```python
m = 0
b = 0
a = 0.001
epochs = 1000
values = []

for i in range(epochs):
    Y_pred = m * x + b
    D_m = (-2/n) * np.sum(x * (y - Y_pred))
    D_b = (-2/n) * np.sum(y - Y_pred)
    m = m - a * D_m
    b = b - a * D_b
    value = (1/n) * sum((y - Y_pred)**2)
    values.append(value)

    if i % 100 == 0:
        print(f"Epoch {i}: Cost after {i} iterations = {value}")
```

```python
print(f"Slope (m) : {np.round(m,2)}")
print(f"Intercept (b) : {np.round(b,2)}")

print(f"y = {np.round(m,2)} * x + {np.round(b,2)}")
```
